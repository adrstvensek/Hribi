## Vrste

V tem podpoglavju bomo pogledali kako visoko se pojavljajo določeni objekti in naravne znamenitosti.

```python
hribi_vrsta = hribi['vrsta']
je_vrh = hribi[hribi_vrsta.str.contains('vrh')]
je_koca = hribi[hribi_vrsta.str.contains('koča')]
je_jezero = hribi[hribi_vrsta.str.contains('jezero')]
je_planina = hribi[hribi_vrsta.str.contains('planina')]
je_bivak = hribi[hribi_vrsta.str.contains('jezero')]
je_slap =hribi[hribi_vrsta.str.contains('slap')]
je_cerkev =hribi[hribi_vrsta.str.contains('cerkev')]

imena_vrst = ['vrh', 'koca', 'jezero', 'planina', 'bivak', 'slap', 'cerkev']
vrste = [je_vrh, je_koca, je_jezero, je_planina, je_bivak, je_slap, je_cerkev]
povprečno_mesto_priljubljenosti=[]
for i,j in zip(vrste, imena_vrst):
    tabela = pd.DataFrame(index=[str(j)], columns = ['povprečno mesto glede na priljubljenost', 'najboljše mesto', 'najslabše mesto'])
    tabela['povprečno mesto glede na priljubljenost'] = sum(i['priljubljenost'])/len(i)
    povprečno_mesto_priljubljenosti.append(tabela)
    
povprečno_mesto_priljubljenosti1 = pd.concat(povprečno_mesto_priljubljenosti)
povprečno_mesto_priljubljenosti1['povprečno mesto glede na priljubljenost'] = povprečno_mesto_priljubljenosti1['povprečno mesto glede na priljubljenost'].round(2)
povprečno_mesto_priljubljenosti1.sort_values('povprečno mesto glede na priljubljenost', ascending = True)

```

```python
števila_ponovitev = []
for i,j in zip(vrste, imena_vrst):
    tabela = pd.DataFrame(index=[str(j)], columns = ['Število'])
    tabela['Število'] = len(i)
    števila_ponovitev.append(tabela)
števila_ponovitev1 = pd.concat(števila_ponovitev)
št_po_vrstah = števila_ponovitev1.sort_values('Število',ascending = False)
št_po_vrstah
```
